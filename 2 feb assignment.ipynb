```python
Q1. Explain with an example each when to use a for loop and a while loop.


For Loop
   - Use a for loop when you know the number of iterations beforehand or when you want to iterate over a sequence (e.g., list, tuple, string).
   - For loops are particularly useful when you have a definite range or sequence to iterate over.

Example:

# Printing numbers from 1 to 5 using a for loop
for i in range(1, 6):
    print(i)

While Loop
   - Use a while loop when you're unsure of the number of iterations beforehand or when you need to repeat a block of code until a certain condition is met.
   - While loops are beneficial when the iteration depends on a condition.

Example:

# Printing numbers from 1 to 5 using a while loop
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
Q2. Write a python program to print the sum and product of the first 10 natural numbers using for
and while loop.

# Using for loop to calculate sum and product of the first 10 natural numbers
n = 10
sum_n = 0
product_n = 1

for i in range(1, n+1):
    sum_n += i
    product_n *= i

print("Using for loop:")
print("Sum of the first", n, "natural numbers:", sum_n)
print("Product of the first", n, "natural numbers:", product_n)


# Using while loop to calculate sum and product of the first 10 natural numbers
n = 10
sum_n = 0
product_n = 1
count = 1

while count <= n:
    sum_n += count
    product_n *= count
    count += 1

print("Using while loop:")
print("Sum of the first", n, "natural numbers:", sum_n)
print("Product of the first", n, "natural numbers:", product_n)
```

```python
Q3. Create a python program to compute the electricity bill for a household.
The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per
unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will
be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.
You are required to take the units of electricity consumed in a month from the user as input.
Your program must pass this test case: when the unit of electricity consumed by the user in a month is
310, the total electricity bill should be 2250.

def calculate_bill(units):
    total_bill = 0
    
    if units <= 100:
        total_bill = units * 4.5
    elif units <= 200:
        total_bill = 100 * 4.5 + (units - 100) * 6
    elif units <= 300:
        total_bill = 100 * 4.5 + 100 * 6 + (units - 200) * 10
    else:
        total_bill = 100 * 4.5 + 100 * 6 + 100 * 10 + (units - 300) * 20
    
    return total_bill

# Taking input from the user
units_consumed = int(input("Enter the units of electricity consumed in a month: "))

# Calculating the electricity bill
bill_amount = calculate_bill(units_consumed)

print("Total electricity bill:", bill_amount, "rupees")
```

```python
Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each
number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print
that list.

# Using for loop to calculate cubes and append numbers divisible by 4 or 5
numbers_list = list(range(1, 101))
result_list = []

for num in numbers_list:
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        result_list.append(num)

print("Numbers whose cube is divisible by 4 or 5 (using for loop):", result_list)


# Using while loop to calculate cubes and append numbers divisible by 4 or 5
numbers_list = list(range(1, 101))
result_list = []

index = 0
while index < len(numbers_list):
    num = numbers_list[index]
    cube = num ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        result_list.append(num)
    index += 1

print("Numbers whose cube is divisible by 4 or 5 (using while loop):", result_list)

```

```python
Q5. Write a program to filter count vowels in the below-given string.
string = "I want to become a data scientist"


string = "I want to become a data scientist"

# Define a function to count vowels
def count_vowels(input_string):
    vowels = 'aeiouAEIOU'
    count = 0
    for char in input_string:
        if char in vowels:
            count += 1
    return count

# Count vowels in the given string
vowel_count = count_vowels(string)

print("Number of vowels in the string:", vowel_count)
```
